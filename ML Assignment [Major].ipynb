# Making Dataset Outliers free

```python
def IQR(col):
    Q3=np.quantile(df[col],0.75)
    Q1=np.quantile(df[col],0.25)
    
    IQR = Q3 - Q1
    
    lower_limit = Q1 - 1.5*IQR
    upper_limit = Q1 + 1.5*IQR
    
    outlier_free = df[((df[col] >= lower_limit) & (df[col] <=upper_limit ))]
    
    df[col] = np.where(df[col] > upper_limit,
        upper_limit,
        np.where(df[col] < lower_limit,
        lower_limit,
        df[col]))
    print(f"The dataset for column {col} is outlier-free.")
    
i=1
for i in range(1,50) :
    print(f"{IQR(f'f_{i}')}")
```

# X, Y value Seperation

```python
X= df.drop(["target"], axis=1)
y= df.target

print(X.shape)
print(y.shape)
```

# Spliting Training and Testing Data using Scikit-Learn

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=.20,random_state=4)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
import math
from sklearn.metrics import f1_score, matthews_corrcoef , classification_report
```

# Building LogisticRegression ML model.

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(x_train, y_train)
y_pred = log_reg.predict(x_test)
A_s= accuracy_score(y_test, y_pred)
print("The Accuracy_Score is :-",{A_s})
pre_scr = precision_score(y_test, y_pred)
print("The Precision_Score  is :-",pre_scr)
rec_scr = recall_score(y_test, y_pred)
print("The Recall_Score is :-",rec_scr)
f1_scr = f1_score(y_test, y_pred)
print("The F1-Score is :-",f1_scr)
Mcc = matthews_corrcoef(y_test, y_pred)
print("The Matthews correlation coefficient is :-",Mcc)
cm=confusion_matrix(y_test, y_pred)
print("Confusion matrix:-")
print(cm)
c_r=classification_report(y_test, y_pred)
print("The Classification_Report is :-",c_r)  
```
